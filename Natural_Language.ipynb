```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
nltk.download('averaged_perceptron_tagger')

my_stop_words = set(stopwords.words('english'))
from textblob import TextBlob
texts = [
    "This is the first document first document is special",
    "This is second document but first document is special"
]
clean_texts = []
for text in texts:
  text = str(TextBlob(text).correct())
  text = text.lower()
  tokenizer = RegexpTokenizer(r"\w+")
  tokens = tokenizer.tokenize(text)
  lemmatizer = WordNetLemmatizer()
  clean_tokens = []
  for token in tokens:
    if token not in my_stop_words:
      token = lemmatizer.lemmatize(token)
      clean_tokens.append(token)

  tagged = pos_tag(clean_tokens)

  clean_text = " ".join(clean_tokens)
  clean_texts.append(clean_text)

model = CountVectorizer()
result = model.fit_transform(clean_texts)

result = result.toarray()
result = pd.DataFrame(result,columns = model.get_feature_names_out())

result
```

```python
import pandas as pd
import nltk
#nltk.download('stopwords')
#nltk.download('wordnet')
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
mystopwords = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

df = pd.read_csv(r"/content/bbc_news.csv")
df = df.head(50)
df.drop(['title','pubDate','guid','link'],axis=1,inplace=True)

# nlp codings

tokenizer = RegexpTokenizer(r'\w+')
lemmatizer =WordNetLemmatizer()
clean_text =[]
for index in df.index:
  news = df.loc[index]['description']
  news= news.lower()
  tokens = tokenizer.tokenize(news)
  new_token =[]
  for token in tokens:
    if token not in mystopwords:
      token = lemmatizer.lemmatize(token)
      new_token.append(token)
  clean_text.append(new_token)
df["clean_text"]= clean_text
data = df['clean_text'].values.tolist()
data



```

```python
id2word = corpora.Dictionary(data)
corpus = [id2word.doc2bow(text) for text in data]
num_of_topics = 2
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_of_topics)
import pyLDAvis.gensim
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(lda_model, corpus,id2word)
```

```python
import pandas as pd
import nltk
#nltk.download('stopwords')
#nltk.download('wordnet')
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
mystopwords = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
df = pd.read_csv(r"/content/bbc_news.csv")
df = df.head(50)
df.drop(['title','pubDate','guid','link'],axis=1,inplace=True)

tokenizer = RegexpTokenizer(r"\w+")
lemmatizer = WordNetLemmatizer()
clean_text = []
for index in df.index:
  news = df.loc[index]['description']
  news = news.lower()
  tokens = tokenizer.tokenize(news)
  news_tokens = []
  for token in tokens:
    if token not in mystopwords:
      token = lemmatizer.lemmatize(token)
      news_tokens.append(token)
  clean_text.append(" ".join(news_tokens))
df['clean_text'] = clean_text
data = df['clean_text'].values.tolist()
data
```

```python
from LeXmo import LeXmo
from textblob import TextBlob
index = 0
result = []
for text in data:
  emotion = LeXmo.LeXmo(text)
  tb = TextBlob(text)
  if (tb.sentiment.polarity > -0.1) and (tb.sentiment.polarity < 0.1):
    sentiment = "neutral"
  elif (tb.sentiment.polarity >=0.1):
    sentiment="Positive"
  else:
    sentiment = "Negative"
  emotion['sentiment'] = sentiment
  result.append(emotion)
result = pd.DataFrame(result)
result
```
